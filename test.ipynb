```python
import plotly.express as px
```

```python
def bar(column):
    df_new = df_controls_1[[column]].value_counts().reset_index()
    df_new.rename(columns={0 : 'Sum of patients'}, inplace=True)
    
    plot = px.bar(df_new, x=column, y='Sum of patients', color=column)
    return plot
```

```python
bar('sex')
```
